```python
import pandas as pd
from financial_analysis import Portfolio
```

```python
risk_free_rate = 0.0016
```

```python
df = (
    pd.read_excel("data.xlsx", sheet_name="etf data", nrows=61)
        .drop(columns="Unnamed: 14")
        .set_index("Date")
)

df
```

```python
market_data = df["^GSPC"]
```

```python
df.drop(columns="^GSPC", inplace=True)
```

```python
portfolio = Portfolio(df, market_data, risk_free_rate)

assert portfolio.annual_risk_free_rate == (risk_free_rate * 12.)
```
